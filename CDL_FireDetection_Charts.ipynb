```python
def calculate_daily_detections(df):
    # Ensure 'acq_date' is in datetime format
    df['acq_date'] = pd.to_datetime(df['acq_date'])
    # Create temporary DataFrame with 'acq_date' as the index
    fires_df = df.set_index('acq_date')
    fires_df['detections'] = 1
    # Resample to get daily detections
    daily_detections = fires_df.resample('D').sum()['detections']
    return daily_detections

# Calculate
daily_detections_2009 = calculate_daily_detections(df_2009)
daily_detections_2010 = calculate_daily_detections(df_2010)
daily_detections_2011 = calculate_daily_detections(df_2011)
daily_detections_2012 = calculate_daily_detections(df_2012)
daily_detections_2013 = calculate_daily_detections(df_2013)
daily_detections_2014 = calculate_daily_detections(df_2014)
daily_detections_2015 = calculate_daily_detections(df_2015)
daily_detections_2016 = calculate_daily_detections(df_2016)
daily_detections_2017 = calculate_daily_detections(df_2017)
daily_detections_2018 = calculate_daily_detections(df_2018)

# Plot line graphs as subplots
fig, axs = plt.subplots(nrows=5, ncols=2, figsize=(15, 12))  # Changed to 5 rows and 2 columns
daily_detections_2009.plot(kind='line', ax=axs[0, 0])
axs[0, 0].set_title('Number of Fire Detections in 2009', fontsize=14, fontweight='bold')
axs[0, 0].set_xlabel('Date', fontweight='bold')
axs[0, 0].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[0, 0].tick_params(axis='both', which='major', labelsize=12)
axs[0, 0].grid()

daily_detections_2010.plot(kind='line', ax=axs[0, 1], color='orange')
axs[0, 1].set_title('Number of Fire Detections in 2010', fontsize=14, fontweight='bold')
axs[0, 1].set_xlabel('Date', fontweight='bold')
axs[0, 1].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[0, 1].tick_params(axis='both', which='major', labelsize=12)
axs[0, 1].grid()

daily_detections_2011.plot(kind='line', ax=axs[1, 0], color='green')
axs[1, 0].set_title('Number of Fire Detections in 2011', fontsize=14, fontweight='bold')
axs[1, 0].set_xlabel('Date', fontweight='bold')
axs[1, 0].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[1, 0].tick_params(axis='both', which='major', labelsize=12)
axs[1, 0].grid()

daily_detections_2012.plot(kind='line', ax=axs[1, 1], color='red')
axs[1, 1].set_title('Number of Fire Detections in 2012', fontsize=14, fontweight='bold')
axs[1, 1].set_xlabel('Date', fontweight='bold')
axs[1, 1].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[1, 1].tick_params(axis='both', which='major', labelsize=12)
axs[1, 1].grid()

daily_detections_2013.plot(kind='line', ax=axs[2, 0], color='purple')
axs[2, 0].set_title('Number of Fire Detections in 2013', fontsize=14, fontweight='bold')
axs[2, 0].set_xlabel('Date', fontweight='bold')
axs[2, 0].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[2, 0].tick_params(axis='both', which='major', labelsize=12)
axs[2, 0].grid()

daily_detections_2014.plot(kind='line', ax=axs[2, 1], color='brown')
axs[2, 1].set_title('Number of Fire Detections in 2014', fontsize=14, fontweight='bold')
axs[2, 1].set_xlabel('Date', fontweight='bold')
axs[2, 1].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[2, 1].tick_params(axis='both', which='major', labelsize=12)
axs[2, 1].grid()

daily_detections_2015.plot(kind='line', ax=axs[3, 0], color='deeppink')
axs[3, 0].set_title('Number of Fire Detections in 2015', fontsize=14, fontweight='bold')
axs[3, 0].set_xlabel('Date', fontweight='bold')
axs[3, 0].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[3, 0].tick_params(axis='both', which='major', labelsize=12)
axs[3, 0].grid()

daily_detections_2016.plot(kind='line', ax=axs[3, 1], color='navy')
axs[3, 1].set_title('Number of Fire Detections in 2016', fontsize=14, fontweight='bold')
axs[3, 1].set_xlabel('Date', fontweight='bold')
axs[3, 1].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[3, 1].tick_params(axis='both', which='major', labelsize=12)
axs[3, 1].grid()

daily_detections_2017.plot(kind='line', ax=axs[4, 0], color='yellowgreen')
axs[4, 0].set_title('Number of \nFire Detections in 2017', fontsize=14, fontweight='bold')
axs[4, 0].set_xlabel('Date', fontweight='bold')
axs[4, 0].set_ylabel('Number of Fire Detections', fontweight='bold')
axs[4, 0].tick_params(axis='both', which='major', labelsize=12)
axs[4, 0].grid()

daily_detections_2018.plot(kind='line', ax=axs[4, 1], color='black')
axs[4, 1].set_title('Number of Fire Detections in 2018', fontsize=14, fontweight='bold')
axs[4, 1].set_xlabel('Date', fontweight='bold')
axs[4, 1].set_ylabel('Number of \nFire Detections', fontweight='bold')
axs[4, 1].tick_params(axis='both', which='major', labelsize=12)
axs[4, 1].grid()

# Hide blank/unused subplots by turning off axis
#for i in range(5):  # 5 subplot rows
 #   for j in range(2):  # 2 subplot cols
  #      if (i == 4 and j > 1):  # Hide subplots in last row after the second column
   #         axs[i, j].axis('off') 
    #    else:
     #       axs[i, j].axis('on') 
    
plt.tight_layout()
plt.show()
```

```python
# Create a mapping dictionary for each year using a dictionary comprehension of LandCoverValueID and corresponding LandCoverTypeID columns, stop before NaN values as excel auto fill under IDs
# Function to create mapping dictionary
def create_mapping_dict(df):
    mapping_dict = {}
    for value_id, value_type in zip(df['LandCoverValueID'], df['LandCoverTypeID']):
        if pd.isnull(value_type):
            break
        mapping_dict[value_id] = value_type
    return mapping_dict

# Create mapping dictionaries for each year
mapping_dicts = {
    2009: create_mapping_dict(df_2009),
    2010: create_mapping_dict(df_2010),
    2011: create_mapping_dict(df_2011),
    2012: create_mapping_dict(df_2012),
    2013: create_mapping_dict(df_2013),
    2014: create_mapping_dict(df_2014),
    2015: create_mapping_dict(df_2015),
    2016: create_mapping_dict(df_2016),
    2017: create_mapping_dict(df_2017),
    2018: create_mapping_dict(df_2018)
}

# Combine mapping dictionaries into one, avoiding duplicates
combined_mapping_dict = {}
for year, mapping in mapping_dicts.items():
    combined_mapping_dict.update(mapping)

#combined_mapping_dict

# Fill in empty LandcoverType entries with the mapping_dict of values and corresponding landcover types
def fill_land_cover_type(row):
    if pd.isnull(row['LandCoverType']):
        return combined_mapping_dict.get(row['LandCoverValue'], "Unknown")
    return row['LandCoverType']

# Apply the filling function to all data frames
all_data = [df_2009, df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018]
for df in all_data:
    df['LandCoverType'] = df.apply(fill_land_cover_type, axis=1)

# All_data is a list of dataframes for each year
land_cover_counts = {year: df['LandCoverType'].value_counts() for year, df in zip(range(2009, 2019), all_data)}
# Convert to DataFrame and fill NaN with zeros
land_cover_counts_df = pd.DataFrame(land_cover_counts).fillna(0)

# make table of below total counts of each landcover type in data frame land_cover_counts_df 
# Create a summary table of total counts for each land cover type
land_cover_totals = land_cover_counts_df.sum(axis=1).reset_index()
land_cover_totals.columns = ['LandCoverType', 'TotalCount']
land_cover_totals = land_cover_totals.rename(columns={'LandCoverType': 'Landcover Type', 'TotalCount': 'Number of MODIS/VIIRS SNPP Fire Detections\n Over Each Landcover'})

# Order the chart from highest to lowest total count
land_cover_totals = land_cover_totals.sort_values(by='Number of MODIS/VIIRS SNPP Fire Detections\n Over Each Landcover', ascending=False)

# Add a total row to the end of the table
total_row = pd.DataFrame({'Landcover Type': ['Total Fire Detections Over Vegetation/Crops'], 
                          'Number of MODIS/VIIRS SNPP Fire Detections\n Over Each Landcover': [land_cover_totals['Number of MODIS/VIIRS SNPP Fire Detections\n Over Each Landcover'].sum()]})
land_cover_totals = pd.concat([land_cover_totals, total_row], ignore_index=True)

# Display the table with outlines
#print(tabulate(land_cover_totals, headers='keys', tablefmt='fancy_grid', showindex=False))

# Create a heatmap of fire detections per landcover type from the above table
# Rows will be individual Landcover Types, colorized from light to dark by Number of MODIS/VIIRS SNPP Fire Detections Over Each Landcover counts column

plt.figure(figsize=(6,16))
# drop last total row calculation from table above from Landcover Type data for heatmap
heatmap_data = land_cover_totals.set_index('Landcover Type').drop('Total Fire Detections Over Vegetation/Crops') 
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".0f", cbar_kws={'label': 'Number of Detections', 'shrink': 0.5})
plt.title('Amount of Midwest Fire Detections at Each Landcover Type', fontweight='bold', fontsize=14)
plt.xlabel('Amount of Fire Detections', fontweight='bold', fontsize=12)
plt.xticks([]) # remove extra x axis label from original column header
plt.ylabel('Landcover Type', fontweight='bold', fontsize=12)
plt.show()
```
